```python
import DataGenerator
import KalmanFilter
import PlotFunctions

from datetime import datetime
import numpy as np
```

## Parameters

```python
date_start = datetime(2022, 1, 1)
date_end = datetime(2022, 7, 31)

initial_value = 6
cov_matrix = 0.4
sampling_variance = 1.5
seed = 40
```

## Generate Time-Series

```python
time_series = DataGenerator.SingleLognormalTimeSeries(
    date_start,
    date_end,
    initial_value,
    cov_matrix,
    sampling_variance, 
    seed
)

time_series.generate_time_series()
```

```python
kf = KalmanFilter.KalmanFilterModel(time_variance=0.01, value_col_name='sampled_value')
forecast_df = kf.apply_kf(time_series.time_series_data, 'ts')
```

```python
PlotFunctions.simple_plot(forecast_df, obs_col='sampled_value')
```
