```python
import numpy as np
import pandas as pd
from src.VolumeAcquisition import LinearVolume, QuadraticVolume, CumulativeLognormalVolume
from src.PlotFunctions import simple_line_plot, multi_variable_plot
```

## Volume as Linear function of CPI

```python
volume_model = LinearVolume(1, 1)
df = pd.DataFrame({'cpi': np.linspace(0.1, 10, 1000)})
df['volume'] = df.apply(lambda x: volume_model.calculate_volume(x['cpi']), axis=1)
simple_line_plot(df, 'cpi', 'volume')
```

## Volume as Quadratic function of CPI

```python
volume_model = QuadraticVolume(1, 1)
df = pd.DataFrame({'cpi': np.linspace(0.1, 10, 1000)})
df['volume'] = df.apply(lambda x: volume_model.calculate_volume(x['cpi']), axis=1)
simple_line_plot(df, 'cpi', 'volume')
```

## Volume as Cumulative Prob. of Lognormal Distribution

#### Example 1

```python
avgs = [ 0.5, 1, 1.5, 2.0, 3]
output_data = []
for avg in avgs:
    volume_model = CumulativeLognormalVolume(avg, 1)
    df = pd.DataFrame({'cpi': np.linspace(0.1, 10, 1000)})
    df['volume'] = df.apply(lambda x: volume_model.calculate_volume(x['cpi']), axis=1)
    df['avg'] = avg
    output_data.append(df)
    
output_data = pd.concat(output_data)
multi_variable_plot(output_data, 'cpi', 'avg', 'volume')
```

#### Example 2

```python
avgs = [ 0.5, 1, 1.5, 2.0]
output_data = []
for avg in avgs:
    volume_model = CumulativeLognormalVolume(avg, .5)
    df = pd.DataFrame({'cpi': np.linspace(0.1, 10, 1000)})
    df['volume'] = df.apply(lambda x: volume_model.calculate_volume(x['cpi']), axis=1)
    df['avg'] = avg
    output_data.append(df)
    
output_data = pd.concat(output_data)
multi_variable_plot(output_data, 'cpi', 'avg', 'volume')
```

```python
simple_line_plot(output_data[output_data['avg'] == 2].reset_index(), 'cpi', 'volume')
```
