## Interpreting that I bid the average value, but the value varies from time to time

```python
from itertools import product

estimated_value_ltv = 5

volume_model = LinearVolume(100, 100)
profit_model = StandardModel(volume_model, 5)


ltv_sample = np.random.normal(estimated_value_ltv, 2, 10000)
fraction_of_estimation = np.linspace(0.3, 1, 100)

df = pd.DataFrame(list(product(ltv_sample, fraction_of_estimation)), columns=['true_ltv', 'ltv_fraction'])
df['cpi'] = estimated_value_ltv * df['ltv_fraction'] 
df['profit'] = df.apply(lambda x: profit_model.calculate_profit(x['cpi'], x['true_ltv']), axis=1)
```

```python
summarized_df = df.groupby(['ltv_fraction'])['profit'].mean().reset_index()
summarized_df.sort_values('profit', ascending=False)
```

## Run for several values of variance

```python
true_value_ltv = 1
sampleIsize = 20000

iter_df = []
# sd_values = [0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4]
sd_values = [0,0.1, 0.2, 0.3, 0.4]


for sd in sd_values:

    volume_model = QuadraticVolume(100, 100)
    profit_model = StandardModel(volume_model, true_value_ltv)


    ltv_sample = np.random.normal(true_value_ltv, sd, sampleIsize)
    fraction_of_estimation = np.linspace(0.3, 1, 20)

    df = pd.DataFrame(list(product(ltv_sample, fraction_of_estimation)), columns=['estimated_ltv', 'ltv_fraction'])
    df['cpi'] = df['estimated_ltv'] * df['ltv_fraction'] 
    df['profit'] = df.apply(lambda x: profit_model.calculate_profit(x['cpi'], true_value_ltv), axis=1)
    df['sd'] = sd
    iter_df.append(df)

```

```python
results_df = pd.concat(iter_df)
summarized_results_df = results_df.groupby(['sd', 'ltv_fraction'])['profit'].mean().reset_index()
summarized_results_df = summarized_results_df.sort_values('profit', ascending=False)
final_results_df = summarized_results_df.groupby('sd')['profit', 'ltv_fraction'].first().reset_index()
```

```python
simple_line_plot(final_results_df, 'sd', 'ltv_fraction')
```
